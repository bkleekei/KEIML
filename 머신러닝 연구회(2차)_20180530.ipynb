```python
import nltk
```

```python
from konlpy.corpus import kolaw
```

```python
from konlpy.tag import Twitter; t = Twitter()
```

```python
ko_con_text = kolaw.open('constitution.txt').read()
ko_con_text
```

```python
tokens_ko = t.nouns(ko_con_text)
tokens_ko
```

```python
len(tokens_ko)
```

```python
stop_words = {'입', '애', '로써', '더욱', '안', '함', '것', '년', '월', '일', '차', '제', '장', '조', '바', '진', '그'}

tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
tokens_ko
```

```python
len(tokens_ko)
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 헌법')
```

```python
ko.vocab().most_common(50)
```

```python
stop_words2 = {'수', '때', '이', '및', '경우', '이상', '범위'}

tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words2]
tokens_ko
```

```python
len(tokens_ko)
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 헌법')
```

```python
ko.vocab().most_common(50)
```

```python
import numpy as np
from PIL import Image
```

```python
korea_coloring = np.array(Image.open("./South_Korea_map_-_ko.png"))

from wordcloud import ImageColorGenerator
image_colors = ImageColorGenerator(korea_coloring)
```

```python
plt.figure(figsize = (12, 12))
plt.imshow(korea_coloring, cmap=plt.cm.gray, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(font_path='/Library/Fonts/AppleGothic.ttf',
                     relative_scaling = 0.2, mask=korea_coloring,
                     background_color='white', 
                     min_font_size = 1, max_font_size = 40
                     ).generate_from_frequencies(tmp_data)
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```
